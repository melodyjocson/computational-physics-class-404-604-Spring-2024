```python
import numpy as np
import matplotlib.pyplot as plt

#initialize 
G=-9.81 #account for gravity 
tend=10  
h=0.01
y=[0,0,0,4]
t=0

def f(t, y):
    f=np.zeros(4)
    r=np.sqrt(y[0]*y[0]+y[2]*y[2])
    f[0]=y[1]
    f[1]=y[2]
    f[2]=y[3]
    f[3]=G
    return f
    
# stays the same 
def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2)
    k4=h*f(t+h, y+k3)
    y += ((k1+2*k2+2*k3+k4)/6)
    return y

yp=[]
for i in np.linspace(0, tend, 1000):
    y = rk4(i, h, y, 4)
    if y[2] < 0:
        break
    yp = np.append(yp, y[2])

plt.plot(np.linspace(0, i, len(yp)),yp)
plt.xlabel('seconds')
plt.ylabel('meters')
plt.show()
```
