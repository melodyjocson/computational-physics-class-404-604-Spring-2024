```python
import numpy as np
import matplotlib.pyplot as plt

#set up x values, as well as x range
x = np.arange(-np.pi, np.pi, .01)

#create a function for reuse 
def equ(x, n):
    y = (np.pi / 2)
    for n in np.arange(1, n+1, 2):
        d = (np.sin(n * x)) / n
        y += d
    return y

#plotting 3 lines using different n values
plt.plot(x, equ(x,11))
plt.plot(x, equ(x,21))
plt.plot(x, equ(x,31))


"""
n = 21
for m in np.arange(3, n+1, 2):
    d = (np.sin(m * x)) / m
    y += d
y = (np.pi / 2) + 2 * (np.sin(x)+ y) / n
plt.plot(x, y)

n = 31
for m in np.arange(3, n+1, 2):
    d = (np.sin(m * x)) / m
    y += d
y = (np.pi / 2) + 2 * (np.sin(x)+ y) / n
plt.plot(x, y)
"""


plt.show()

```
