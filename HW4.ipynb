```python
#1:

# libraries 
import numpy as np 
import matplotlib.pyplot as plt

#forward difference 
def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

#change x to pi/4
x= np.pi/4 

dxarr=[]
yarr=[]
for dx in 10**np.arange(-18, 0.1, 0.1):
    y=derifw(np.cos,x,dx)
    #print('dx %g,  y %g'%(dx,y))
    dxarr.append(dx)
    yarr.append(y)
    
#plotting 
plt.plot(dxarr, abs(yarr-(-np.sin(x))), 'b-', lw=2) 
#set plots to show log
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
At 10e16, we can see the line starts to decrease at n = 1. The optimal value for dx is approx 10e-19. 
```

```python
#2:


#central difference
def dericd(y,x,dx):
    return (y(x + (dx/2)) - y(x - (dx/2)))/dx

#change x to pi/4
x= np.pi/4 

dxarr=[]
yarr=[]
for dx in 10**np.arange(-17, 0.1, 0.1):
    y=dericd(np.cos,x,dx)
    #print('dx %g,  y %g'%(dx,y))
    dxarr.append(dx)
    yarr.append(y)
    
#plotting 
plt.plot(dxarr, abs(yarr-(-np.sin(x))), 'b-', lw=2) 
#set plots to show log
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
The central difference scheme is better because the forward difference due to the trucation error being smaller in
central difference compared to the forward. 
```
