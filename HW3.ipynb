```python
#libraries
import time
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
"""i would like to note that this program does not work sometimes
in jupyter, but it works in visual studio code
"""
# Recursion method
print("Recusion Method")
def fib2(n):#function for recursion
    if n == 0: return 0 
    elif n == 1: return 1 #excecutes program when error occurs 
    else: return fib2(n-1)+fib2(n-2) #returns the recusion function 

n1 = np.arange(1, 21 ,1) #range for a set of n numbers 
time1=np.zeros(n1.size) 

for n in range(n1.size): #loop fib2 function for a set of n numbers 
    start_time = time.time() #sets a timer for each iteration 
    temp1=fib2(n1[n]) # result of function at each iteration
    time1[n] = time.time()-start_time
    print("Iteration",n+1) #shows which iteration 
    print("n =", n) #shows the value of n for an iteration
    print(time1[n], "seconds")  #shows the time for iteration
    print(" ")

plt.plot(np.log10(n1), np.log10(time1), label= 'Recursion Method')

print("Tabulation Method")
def fib4(n):
    table =[0]*(n+1)
    table[1] = 1
    for i in range(2,n+1):
        table[i]=table[i-1]+table[i-2] 
    return table[n]


n2 = np.arange(1,1000 ,100) #range for a set of n numbers 
time2=np.zeros(n2.size) 
for i in range(n2.size):
    start_time = time.time()
    temp1=fib4(n2[i]) # result of function at each iteration
    time2[i] = time.time()-start_time
    print("n =", i) #shows the value of n for an iteration
    print(time2[i], "seconds")
    print(" ")
    

plt.plot(np.log10(n2), np.log10(time2), label='Tabulation Method')

#Matrix Method
print("Matrix Method")
def arrpow(arr, n):
    yarr=arr
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr
def fib5(n):
    arr= [[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]
    return f

x = 0
n3 = np.arange(1, 10000, 1)
time3 = np.zeros(n3.size) 
start_time = time.time()

for n in np.arange(1,1000,1):
    time3[n] = time.time() - start_time
    temp3=fib5(n+1) # result of function at each iteration
    print("Iteration", x+1)
    print("n =", n)
    print(time3[n], "seconds")
    x+=1

plt.plot(np.log10(n3), np.log10(time3), label='Matrix Method')
plt.legend()
plt.show()
```

```python
Using the Reursion method, we can see that the larger n is,
the longer it takes to generate a result. We can note that 
depending on each run of the program, we can see that there are 
multiple points in the graph wheconsistentre a larger n can sometimes
run quicker than it's previous iteration. Ultimately, n will get 
larger enough to the point that it exponetially takes longer after
each iteration. This method eventually will stop working after 
a set of iterations, making its last iterations taking siginficantly
longer than other methods. 

With the Tabulation method, results have shown to be more consistent,
as n gets larger it slowly increases the time it takes to get results
after each iteration. Again, there are some instances where a value
of n (much less than recursion method, though) will generate 
slightly faster than it's previous iteration. 

For the matrix method, results are more consistent(no iteration is
slower than it's predecessor). Generally, it takes slightly 
longer to produce results than the tabulation method.                                                    
```
